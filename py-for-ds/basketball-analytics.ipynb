## Function with basic visualization

```python
def myplot_basic(playerlist):
    
    for name in playerlist:
        plt.plot(Games[Pdict[name]], c='Black', ls='--', marker='s', ms=7, label=name)
        
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation='horizontal')
    plt.show()
```

```python
myplot_basic(['KobeBryant','LeBronJames', 'DerrickRose'])
```

## Fixed function with expanded visualization

```python
def myplot_adv(data, playerlist=Players):
    
    Clr = {"KobeBryant":"Black", "JoeJohnson":"Red", "LeBronJames":"Green", "CarmeloAnthony":"Blue", "DwightHoward":"Magenta", "ChrisBosh":"Black", "ChrisPaul":"Red", "KevinDurant":"Green", "DerrickRose":"Blue", "DwayneWade":"Magenta"}
    Mrk = {"KobeBryant":"s", "JoeJohnson":"o", "LeBronJames":"^", "CarmeloAnthony":"D", "DwightHoward":"s", "ChrisBosh":"o", "ChrisPaul":"^", "KevinDurant":"D", "DerrickRose":"s", "DwayneWade":"o"}
    
    for name in playerlist:
        plt.plot(data[Pdict[name]], c=Clr[name], ls='--', marker=Mrk[name], ms=7, label=name)
        
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation='horizontal')
    plt.show()
```

```python
myplot_adv(Games, ['KobeBryant','LeBronJames', 'DerrickRose'])
```

```python
myplot_adv(Salary, ['KobeBryant', 'LeBronJames', 'DerrickRose'])
```
