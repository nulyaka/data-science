```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import math

x = sp.Symbol('x')
```

### 1.a

```python
# Derivatives in the book are not presented yet
# Usage of simple functions

def get_dst(t):
    dst = 0
    
    if t <= 2:
        dst = 30*t
    elif t >= 2 and t <= 4:
        dst = 60
    elif t >= 4 and t <= 6:
        dst = 30*(6-t)
        
    return dst

def get_vel(t):
    vel = 0
    
    if t < 2:
        vel = 30
    elif t >= 4 and t <= 6:
        vel = -30
        
    return vel
```

```python
# 200 linearly spaced numbers
Time = np.linspace(0,6,200)
Dst = [get_dst(t) for t in Time]
Vel = [get_vel(t) for t in Time]

plt.plot(Time, Dst)
plt.show()
plt.plot(Time, Vel)
plt.show()
```

### 1.b

```python
def get_dst(t):
    dst = 20*t-20
    
    if t < 1:
        dst = 10 - 10*t
    
    return dst

def get_vel(t):
    vel = 20
    
    if t <= 1:
        vel = -10
        
    return vel
```

```python
def U(t):
    if t < 0:
        return 0
    
    return 1  

Points = np.linspace(-10,10,500)
A = [U(p) for p in Points]
B = [a+2 for a in A]
C = [U(p+2) for p in Points]
D = [3*a for a in A]
E = [U(3*p) for p in Points]

plot1 = plt.figure(1)
plt.plot(Points,A,c='cyan')
plt.plot(Points,B,c='green')

plot2 = plt.figure(2)
plt.plot(Points,C,c='black')
plt.plot(Points,D,c='magenta')

plt.figure(3)
plt.plot(Points,E,c='red')

plt.show()
```

```python
def f(x):
    return pow(x,2)+x


X = np.linspace(-5,5,1000)
A = [f(x) for x in X]
B = [f(a) for a in A]
C = [f(x+1) for x in X]

plt.figure(1)
plt.title('f(x)')
plt.plot(X,A)

plt.figure(2)
plt.title('f(f(x))')
plt.plot(X,B)

plt.figure(3)
plt.title('f(x+1)')
plt.plot(X,C)

plt.show()
```

### 51

```python
def f(x):
    return 3*x-1

FROM = 0
TO = 2
STEPS = 10

line = np.linspace(FROM, TO, STEPS)
a = [f(point+2) for point in line]
b = [f(point)+2 for point in line]
c = [f(point)*2 for point in line]
d = [f(point*2) for point in line]
e = [f(point)*(-1) for point in line]
f = [f(f(point)) for point in line]

plt.figure(1)
plt.plot(line,a)
plt.plot(line,b)
# slope and intercept from least squares
slope1 = np.polyfit(line, a,1)
slope2 = np.polyfit(line, b,1)
print(slope1, slope2)

plt.figure(2)
plt.plot(line,c)
plt.plot(line,d)
slope1 = np.polyfit(line, c,1)
slope2 = np.polyfit(line, d,1)
print(slope1, slope2)

plt.figure(3)
plt.plot(line,e)
plt.plot(line,f)
slope1 = np.polyfit(line, e,1)
slope2 = np.polyfit(line, f,1)
print(slope1, slope2)

plt.show()
```
